## Aim : Use RANSAC to remove outlier key point matches and fit a transformation model between two images.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load image in grayscale
img = cv2.imread('/content/image1.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)
```

```python
# Apply Harris Corner detection
dst = cv2.cornerHarris(gray, blockSize=2, ksize=3, k=0.04)
```

```python
# Dilate the result for marking the corners
dst = cv2.dilate(dst, None)
```

```python
# Threshold for an optimal value (may need tuning)
img[dst > 0.01 * dst.max()] = [0, 0, 255]  # Red corners
```

```python
# Show using matplotlib
plt.figure(figsize=(8, 6))
plt.imshow(img)
plt.title('Harris Corners')
plt.axis('off')  # Hide axes
plt.show()
```
